## Ouverture et traitement du fichier texte

Pour cela, on utilise un fichier texte pré-traité, donc débruité au minimum, correspondant à une page de répertoire de notaire.

```python
with open('../../corpus_test/lectaurep/preprocess_page_for_entityfishing/entry_FRAN_0025_0149_L-1.txt', 'r', encoding='utf-8') as fh:
    sample = fh.read()
```

On utilise le caractère `¬` utilisé pour la segmentation artificielle des enregistrements pour créer une liste où chaque index est un enregistrement.

```python
entries = sample.split('¬')
entries_without_linebreaks = [entry.strip('\n') for entry in entries]
```

On supprime le dernier index vide.

```python
entries_without_linebreaks = entries_without_linebreaks[:-1]
```

## Import du client

Après avoir installé la librairie, on importe le client.

```python
import nerd
```

```python
client = nerd_client.NerdClient()
```

On crée un dictionnaire qui permettra de stocker les réponses HTTP de l'API d'Entity-fishing sous le format suivant :

`{numéro de l'enregistrement: (réponse HTTP d'Entity-fishing)}`

```python
counter = 1
entityfishing_response = {}
for entry in entries_without_linebreaks:
    print(f'Enregistrement numéro {counter}')
    entityfishing_response[f'Enregistrement numéro {counter}'] = client.disambiguate_query(entry, language='fr')
    print('\n')
    counter += 1
```

On peut visualiser l'intégralité des réponses HTTP : 

```python
entityfishing_response
```

Et on peut enfin visualiser les résultats qui nous intéressent :

```python
for key in entityfishing_response:
    print(key)
    print('Entités nommées extraites : \n')
    print(entityfishing_response[key][0]['entities'])
    print('\n')
```
